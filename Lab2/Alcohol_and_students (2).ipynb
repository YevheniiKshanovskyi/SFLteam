1. Do students whose parents live apart consume more alcohol than students whose parents live together?

```python
# ANSWER HERE:
a_mean = [data[data.Pstatus == 'A'].Dalc.mean(), data[data.Pstatus == 'A'].Walc.mean()]
t_mean = [data[data.Pstatus == 'T'].Dalc.mean(), data[data.Pstatus == 'T'].Walc.mean()]
apart = np.mean(a_mean)
together = np.mean(t_mean)
print(f'Index of alcohol consumption for students in living apart families is {round(apart, 3)}, and {round(together, 3)} for living together ones.')
print(f'Technically the answer is {apart > together}, but actually the difference for indecies is too small to matter.')
```

2. Most large families are rural ones?

```python
# ANSWER HERE:
fam = data.groupby(by=['address', 'famsize']).describe()
gt3_in_u = (fam.iat[2,0] / (fam.iat[2,0] + fam.iat[3,0]))*100
gt3_in_r = (fam.iat[0,0] / (fam.iat[0,0] + fam.iat[1,0]))*100
print('Percentage of large families in countryside - ', round(gt3_in_r, 2))
print('Percentage of large families in urban area- ', round(gt3_in_u, 2))
print(f'The answer is {gt3_in_r > gt3_in_u}.')
```

3. To what extent the lack of Internet at home affects the students' grades

```python
# ANSWER HERE:
with_int =  data[data.internet == 'yes'][['G1','G2', 'G3']].mean().mean()
without_int = data[data.internet == 'no'][['G1','G2', 'G3']].mean().mean()
print('Avarage mark with internet at home:\n', round(with_int, 2))
print('Avarage mark with internet at home:\n', round(without_int, 2))
print('Difference here is tiny to have matter too.')
```

4. To what extent does drinking alcohol on weekends affect missed classes

```python
# ANSWER HERE:
alco_and_absence = len(data[(data.Walc > data.Walc.mean()) & (data.absences > data.absences.mean())]) / len(data) * 100
print(f'In {round(alco_and_absence, 2)}% of cases higher alcohol consumption on weekends is accompanied with increased number of absenteeism')
```

5. What is the index of alcohol consumption on weekdays for students, one of whose parents works in the healthcare field

7. What is percentage of students with the strongest health, who attend extra-curricular activities 

```python
# ANSWER HERE:
filtered = data[ (data['health'] == 5) & (data['activities'] == 'yes')]
print('All students\n', data[['health', 'activities']], '\nnumber of the students:', len(data))
print('Students with the strongest health who participated in the events\n', filtered[['health', 'activities']], '\nnumber of the students:', len(filtered))
perc = len(filtered) / len(data) * 100
print('\nPercentage of such students:', round(perc, 2), '%.')
```

8. What percentage of families with more than 3 children have the best family relationships?

```python
# ANSWER HERE:
best = len(data[(data.famrel == 5 ) & (data.famsize == 'GT3')]) / len(data[data.famsize == 'GT3'])*100
worst = len(data[(data.famrel == 1 ) & (data.famsize == 'GT3')]) / len(data[data.famsize == 'GT3'])*100
print(f'The percentage of happy families is {round(best-worst, 2)}% higher then the unhappy one.')
```

9. What is the percentage of students with family relationships <= 2 is not going to take higher education?

10. What is the percentage of students, who consume alcohol over the avarage, have failed exams at least once?
