```python
import pandas as pd
df = pd.read_csv("Euro_2012_stats_TEAM.csv")
df.head()
```

```python
df['total shots'] =  (df['Shots on target'] + df['Shots off target'])
df['blocked own shots'] = (df['Total shots (inc. Blocked)'] - df['total shots'])
df['Block rate%'] = ((df['Blocks'] / df['blocked own shots'])*100)
block_ratings = df[['Team', 'Block rate%', 'Goals', 'Goals conceded']]
block_ratings.sort_values(by='Block rate%', ascending=False).head(5)
```

```python
df['Touches without pass'] = df['Touches'] - (df['Passes']*1.2 + df['Saves made']*2)
df['Touches per dribble'] = df['Touches without pass'] / df['Dribbles']
df['Team Work'] = (df['Passes completed'] / df['Touches per dribble']) / 100
Dribble_stats = df[['Team','Goals', 'Touches per dribble', 'Passes completed', 'Team Work']].sort_values(by='Team Work', ascending=False)
Dribble_stats.head(5)
```

```python
df['Key Passes Rate'] =  df['Passes'] / (df['Shots on target'] - df['Goals'] + df['Penalty goals'] + df['Penalties not scored'])
df['Key Passes'] = round((df['Passes'] /  df['Key Passes Rate']) * (round(df['% Goals-to-shots'].apply(lambda x: float(x.split()[0].replace('%', ''))).mean(),-1)),1)
Key_passes = df[['Team', 'Goals', 'Key Passes']]
Key_passes.sort_values(by='Key Passes', ascending = False).head(5)
```
