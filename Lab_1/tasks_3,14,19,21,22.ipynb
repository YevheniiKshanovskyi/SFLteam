##  Question 3

Write a function that asks the user to enter a list of integers. Do the following:

- Print the last item in the list.
- Print the list in reverse order.
- Print Yes if the list contains a 5 and No otherwise.
- Print the number of fives in the list.
- Remove the first and last items from the list, sort the remaining items, and print the result.
- Print how many integers in the list are less than 5.

### Answer

```python
def func(listt):

  print(listt[-1])

  print(list(reversed(listt)))

  print(5 in listt)

  print(listt.count(5))

  new_list = listt[:]
  new_list.pop(0)
  new_list.pop(-1)
  new_list.sort()
  print(new_list)

  count = 0
  for i in listt:
    if i < 5:
      count += 1
  print(count)



func([1, 2, 4, 3, 5, 6, 7, 8, 9, 10])
```

##  Question 14

Write a program that lets the user play Rock-Paper-Scissors against the computer. There should be three rounds, and after those three rounds, your program should print out who won and lost or that there is a tie.

### Answer

```python
from random import *


def play():
  '''plays game and calculates who's a winner'''

  user_score = 0
  comp_score = 0

  for i in range(3):
    
    user_choice = input('Choice rock, paper or scissors:')
    comp_choice = choice(['rock', 'paper', 'scissors'])

    if user_choice == 'rock' and comp_choice == 'scissors':
      user_score += 1
    elif user_choice == 'rock' and comp_choice == 'paper':
      comp_score += 1
    elif user_choice == 'paper' and comp_choice == 'rock':
      user_score += 1
    elif user_choice == 'paper' and comp_choice == 'scissors':
      comp_score += 1
    elif user_choice == 'scissors' and comp_choice == 'paper':
      user_score += 1
    elif user_choice == 'scissors' and comp_choice == 'rock':
      comp_score += 1
      
    print(f'User: {user_choice}, computer: {comp_choice} \n')

  if user_score > comp_score:
    return f'User won with score {user_score} out of 3.'
  elif user_score < comp_score:
    return f'Computer won with score {comp_score} out of 3.\
 User lost.'
  else:
    return f'There\'s a tie!'

play()
```

## Question 19

Sort a tuple of tuples by 2nd item. For example, the output for the tuple $(('a', 23),('b', 37),('c', 11), ('d',29))$ should be $(('c', 11), ('a', 23), ('d', 29), ('b', 37))$

## Answer

```python
def tuple_sort(tup):
  listt = list(tup)
  listt.sort(key = lambda x: x[1])
  tup_new = tuple(listt)
  return tup_new

tup = (('a', 23), ('b', 37), ('c', 11), ('d', 29))

tuple_sort(tup)
```


## Question 21

Calculate number of distinct characters in a string using a for loop.

## Answer

```python
def calculate_char(string):
  res = []
  for i in string:
    res.append(i)
  dist = len(res)
  return f'distinct characters: {dist}'

string = 'abccdefg'

calculate_char(string)
```

## Question 22

Write a function called is_sorted that is given a list and returns True if the list is sorted and False otherwise.

```python
def is_sorted(listt):
  sort = sorted(listt)
  if listt == sort:
    return True
  else:
    return False

listt = ['a', 'b', 'a']
is_sorted(listt)
```
