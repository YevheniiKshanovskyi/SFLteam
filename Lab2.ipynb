8. На скільки більший відсоток найщасливіших багатодітних сімей від найменш щасливих ( якщо брати від загальної кількості багатодітних)?


```python
best = len(data[(data.famrel == 5 ) & (data.famsize == 'GT3')]) / len(data[data.famsize == 'GT3'])*100
worst = len(data[(data.famrel == 1 ) & (data.famsize == 'GT3')]) / len(data[data.famsize == 'GT3'])*100
print('відсоток найщасливіших багатодітних сімей більший від найменш щасливих на ',round(best-worst,2), '%, якщо брати від загальної кількості багатодітних')
```
